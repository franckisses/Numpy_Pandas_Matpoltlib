```python
tips = pd.read_csv('./data/tips.csv')
```

```python
tips.head()
```

```python
# bin
fig2 = plt.figure()
axes5 = fig2.add_subplot(1,1,1)
axes5.hist(tips['total_bill'] ,bins=10)
axes5.set_title('Histogram of Total Bill')
axes5.set_xlabel('Total Bill')
axes5.set_ylabel('Frequency')

fig2.show()
```

```python
# bivariate

scatter_plot = plt.figure()
axes6 = scatter_plot.add_subplot(1,1,1)
axes6.scatter(tips['total_bill'], tips['tip'])
axes6.set_title('Scatterplot of Total Bill vs Tip')
axes6.set_xlabel('Total Bills')
axes6.set_ylabel('Tips')
scatter_plot.show()
```

```python
# boxplot
boxplot = plt.figure()
axes7 = boxplot.add_subplot(1,1,1)
axes7.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],
    tips[tips['sex'] == 'Male']['tip']],
    labels = ['female', 'male']
)
axes7.set_xlabel('Sex')
axes7.set_ylabel('Tips')
axes7.set_title('Boxplot of Tips by Sex')
boxplot.show()
```

```python
# multi variable 
def record_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
    
tips['sex_color'] = tips['sex'].apply(record_sex)

scatter_plot = plt.figure()
axes8 = scatter_plot.add_subplot(1,1,1)
axes8.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size'] * 10,
    alpha=0.5
)
axes8.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes8.set_xlabel('Total Bill')
axes8.set_ylabel('Tip')
scatter_plot.show()
```

```python
hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
plt.show()
```

```python
hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram')
ax.set_ylabel('Total Bill')
ax.set_xlabel('Frequency')
plt.show()
```

```python
# core density
den, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
plt.show()
```

```python
# rug plot 
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
plt.show()
```

```python
# count 
count, ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
plt.show()
```

```python
# regplot
scatter, ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()
```

```python
fig = sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
joint = sns.jointplot(x='total_bill', y='tip',data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
hexbin =sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
kde, ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], data2=tips['tip'],shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()
```

```python
kde_join = sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
bar,ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
plt.show()
```

```python
box, ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
plt.show()
```

```python
vlolin, ax = plt.subplots()
ax =sns.violinplot(x='time', y='total_bill',data=tips)
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total of bill')
plt.show()
```
